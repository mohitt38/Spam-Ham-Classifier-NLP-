```python
import pandas as pd
import numpy as np
messages=pd.read_csv('SMSSpamCollection', sep='\t', names=['label','messages'])
```

```python
messages
```

```python
## DATA CLEANING AND PREPROCESSING
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]
for i in range (0,len(messages)):
    review=re.sub('[^a-zA-Z]',' ',messages['messages'][i])
    review=review.lower()
    review=review.split()
    review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
## Output Features
y=pd.get_dummies(messages['label'])
y=y.astype(int)
y=y.iloc[:,0]
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(corpus,y,test_size=0.20)
```

```python
## Create the TFID model
from sklearn.feature_extraction.text import TfidfVectorizer
tfid=TfidfVectorizer(max_features=2500,ngram_range=(1,2))
```

```python
len(X_train),len(y_train)
```

```python
# Independent Features
X_train=tfid.fit_transform(X_train).toarray()
X_test=tfid.transform(X_test).toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_detect_model=MultinomialNB().fit(X_train,y_train)
```

```python
y_pred=spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
cs=classification_report(y_test,y_pred)
```
